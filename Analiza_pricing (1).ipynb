## UPLOAD csv fajli sa podacima za 1.1-28.5.2020 period i 1.1.-28.5.2021 period

```python
df20 = pd.read_csv('2020.csv', sep=";")
```

```python
df20
```

#### izbacivanje redaka sa nepostojećim vrijednostima

```python
df20 = df20[df20['Category'].notna()]
```

#### kreiranje indexa od dva ključa (kolone) datum i kategorija

```python
df20 = df20.set_index(['Year.Month No', 'Category'])
```
