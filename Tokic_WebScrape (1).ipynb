```python
import requests
from bs4 import BeautifulSoup

URL = 'https://popusti.njuskalo.hr/f/tokic-poslovnice/grad-zagreb'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

## Nakon pregleda web stranice s F12 u Chromeu
## Definiranje rezultata kao pronalazak spremnika u html-u koji sadrži sve rezultate pretraživanja

```python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://i.imgur.com/AslFePr.png")
```

```python
results = soup.find("ul",{"class":"storeListType2"})
```

```python
print(results.prettify())
```

### Definiranje funkcije poslovnice za podjelu rezultata na pododjeljke

```python
poslovnice = results.find_all('h4', class_='storeMainInfo')
```

```python
for poslovnica in poslovnice:
    print(poslovnica, end='\n'*2)
```

### Definiranje elemenata; naslov i adresa

```python
for poslovnica in poslovnice:
    title_elem = poslovnica.find('span', class_='title')
    address_elem = poslovnica.find('address')
    print(title_elem)
    print(address_elem)
    print()
```

### clean the HTML using .text extension on elements and print results

```python
for poslovnica in poslovnice:
    title_elem = poslovnica.find('span', class_='title')
    address_elem = poslovnica.find('address')
    print(title_elem.text)
    print(address_elem.text)
    print()
```

### uvoz biblioteke pandas za stvaranje podatkovnog okvira iz rezultata pretraživanja

```python
import pandas as pd
```

```python
results = soup.find("ul",{"class":"storeListType2"})
```

### stvaranje 2 odvojena popisa; Ime i adresa koje se popunjava sa svim rezultatima pretraživanja za kreirane elemente

```python
Ime=[]
adresa=[]
for poslovnica in poslovnice:
    title_elem = poslovnica.find('span', class_='title')
    address_elem = poslovnica.find('address')
    Ime.append(title_elem.text)
    adresa.append(address_elem.text)
```

### kreiranje podatkovnog okvira iz popisa imena i adresa

```python
df = pd.DataFrame({'Ime poslovnice':Ime,'Adresa':adresa})
```

```python
df
```

### spremanje podatkovnog okvira u .csv datoteku
(prema zadanom python će se spremiti u csv s razdjelnikom zarezom, u funkciji spremanja možemo promijeniti postavke razdjelnika)

```python
df.to_csv('tokic_poslovnice.csv', index=False, encoding='utf-8')
```
